```python
import sys
import math
from math import sqrt
from sympy.solvers import nonlinsolve
from sympy.solvers import solve
from sympy import symbols

infile1 = open("/home_b/gargi/project_1/attach_sugar_mucin/muc2_smog20_head.gro", "r")
infile2 = open("/home_b/gargi/project_1/attach_sugar_mucin/attach_sug.txt", "r")
#outfile1 = open("/home_b/gargi/project_1/attach_sugar_mucin/sug_coord.txt", "w")
outfile2 = open("/home_b/gargi/project_1/attach_sugar_mucin/muc2_sugar_gro.txt", "w")
outfile3 = open("/home_b/gargi/project_1/attach_sugar_mucin/muc2_sugar_def.txt", "w")
outfile4 = open("/home_b/gargi/project_1/attach_sugar_mucin/muc2_sugar_bond.txt", "w")

res_num = []
x_cord_prot = []
y_cord_prot = []
z_cord_prot = []
count_line = 0

for lines in infile1:
    a, b, c, d, e, f = lines.split()
    a1 = a[0:-3]
    res_num.append(int(a1))
    x_cord_prot.append(float(d))
    y_cord_prot.append(float(e))
    z_cord_prot.append(float(f))
    count_line = count_line + 1

#for lines in infile1:
#    a, b, c, d, e, f, g, h, i = lines.split()
#    res_num.append(int(b))
#    x_cord_prot.append(float(g))
#    y_cord_prot.append(float(h))
#    z_cord_prot.append(float(i))
#    count_line = count_line + 1
    
res_prot_s = []
count_prot_s = 0

for lines2 in infile2:
    res_prot_s.append(int(lines2))
    count_prot_s = count_prot_s + 1
    
#print(count_line, count_prot_s)
x,y,z = symbols('x, y, z', real=True)

for i in range(count_prot_s):
    for j in range(count_line):
        if res_prot_s[i] == res_num[j] and res_prot_s[i] <= 199 and res_prot_s[i] >= 2:
            x2 = x_cord_prot[j]
            y2 = y_cord_prot[j]
            z2 = z_cord_prot[j]
            x3 = x_cord_prot[j+1]
            y3 = y_cord_prot[j+1]
            z3 = z_cord_prot[j+1]
            x33 = x_cord_prot[j-1]
            y33 = y_cord_prot[j-1]
            z33 = z_cord_prot[j-1]
            vec1 = math.sqrt(((x2 - x3)*(x2 - x3)) + ((y2 - y3)*(y2 - y3)) + ((z2 - z3)*(z2 - z3)))
            vec2 = math.sqrt(((x2 - x33)*(x2 - x33)) + ((y2 - y33)*(y2 - y33)) + ((z2 - z33)*(z2 - z33)))
            mult1 = vec1*math.cos((62.8155*math.pi)/180)*0.37783
            mult2 = vec2*math.cos((88.0857*math.pi)/180)*0.37783
            eq1 = ((x - x2)**2 + (y - y2)**2 + (z - z2)**2) - 0.143
            eq2 = ((x - x2)*(x2 - x3)) + ((y - y2)*(y2 - y3)) + ((z - z2)*(z2 - z3)) - mult1
            eq3 = ((x - x2)*(x2 - x33)) + ((y - y2)*(y2 - y33)) + ((z - z2)*(z2 - z33)) - mult2
            sol1, sol2 = nonlinsolve([eq1, eq2, eq3], [x,y,z])
            s1 = float(sol1[0])
            s2 = float(sol1[1])
            s3 = float(sol1[2])
            #ss = 'HETATM{:>5s}  C1  0VA {:>5s}  {:>10.1f} {:>7.1f} {:>7.1f}{}'.format(str(i), str(i), s1, s2, s3, "\n")
            k = i + 40393
            #print(k, s1, s2, s3)
            ss2 = '{:>5d}0VA     C1{:>5d} {:>7.3f} {:>7.3f} {:>7.3f}{}'.format(k, k, s1, s2, s3, "\n")
            ss3 = ' {:>5d}       NB_2  {:>5d}    0VA     C1  {:>5d}{}'.format(k, k, k, "\n")
            outfile2.writelines(ss2)
            outfile3.writelines(ss3)
            #outfile4.writelines(ss4)
            
outfile2.close()
infile1.close()
infile2.close()
    
```
